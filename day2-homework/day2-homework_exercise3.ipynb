# Import Libraries

```python
import pandas as pd
df_SNP = pd.read_csv("/Users/cmdb/Downloads/matrix_1kg.txt",
                    sep = "\t")
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

# Solve for Allelic Frequency

```python
df_SNP_edit = df_SNP.iloc[: , 4:] #POS column interfering with results

total_chromosomes = (2552 - 4) * 2 
counts = df_SNP_edit.sum(axis=1) 

allele_frq = counts/total_chromosomes
```

# Adding "Common Variation"

```python
df_SNP['Allelic Frequency'] = allele_frq

SNP1 = df_SNP.loc[:,'Allelic Frequency'] > 0.05
SNP2 = df_SNP.loc[:,'Allelic Frequency'] < 0.95

SNP3 = SNP1 & SNP2

common_variation = df_SNP.loc[SNP3,:]

common_variation
```
