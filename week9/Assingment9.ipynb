## Working with the HiC data Chr13

```python
hic=hifive.HiC('hic_project_output', 'r')
data_chr13 = hic.cis_heatmap('chr13', 1000000, datatype='fend', arraytype='full', diagonalincluded=True)
```

```python
for value in data_chr13:
    value += 1
data_chr13
```

```python
import numpy as np
```

```python
enrichment_chr13 = np.log2(data_chr13[:, :, 0] / data_chr13[:, :, 1])
```

```python
import matplotlib.pyplot as plt


```

```python
plt.imshow(enrichment_chr13)
plt.show()
```
