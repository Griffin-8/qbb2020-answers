## Violin Plot of A and B compartments

```python
labels = ['A','B']

## combine these different collections into a list
data_to_plot = [fpkm_pos, fpkm_neg]

# Create a figure instance
fig = plt.figure()

# Create an axes instance
ax = fig.add_axes([0,0,1,1])
ax.set_xticklabels(labels)
ax.xaxis.set_ticks_position('bottom')
ax.set_xticks(np.arange(1, len(labels) + 1))
# Create the boxplot
bp = ax.violinplot(data_to_plot)
plt.show()
```

```python
import pyBigWig
bw = pyBigWig.open('data/WT_H3K27me3.bw')
bw.stats('chr1', 1100000, 1200000, type='sum')
```

```python
pos_meth_list = []
pos_expression_list = []
file = 'positives_fpkm.bed'
for line in open(file):
    field = line.strip('\r\n').split('\t')
    start_pos=int(field[1])
    end_pos=int(field[2])
    expression=field[4]
    pos_expression_list.append(expression)
    meth_sum = bw.stats('chr13', start_pos, end_pos, type='sum')
    if meth_sum == [None]:
        pos_meth_list.append([1])
    else:
        pos_meth_list.append(meth_sum)
```

```python
neg_meth_list = []
neg_expression_list = []
file = 'negatives_fpkm.bed'
for line in open(file):
    field = line.strip('\r\n').split('\t')
    start_pos=int(field[1])
    end_pos=int(field[2])
    expression=field[4]
    neg_expression_list.append(expression)
    meth_sum = bw.stats('chr13', start_pos, end_pos, type='sum')
    if meth_sum == [None]:
        neg_meth_list.append([1])
    else:
        neg_meth_list.append(meth_sum)
```

```python
pos_meth_list
```

## A compartment

```python
plt.scatter(pos_expression_list, np.log(pos_meth_list), alpha = 0.5)
plt.show()
```

## B compartment

```python
plt.scatter(neg_expression_list, np.log(neg_meth_list), alpha = 0.5)
plt.show()
```
