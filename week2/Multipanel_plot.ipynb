# Allele Frequency Data

```python
open_file = open('/Users/cmdb/qbb2020-answers/week2/Step7_2.vcf')
counter = 0
three_nine = []

for line in open_file: #less -S to open in terminal 
    if line.startswith('##'):
        continue
    if line.startswith('#'):
        continue
    fields = list(line.strip().split())
    for line in fields:
        subset = fields[-12]
        str1 = ''.join(subset)
        fields2 = list(str1.split(';'))
        three_nine1 = fields2[3]
    three_nine.append(three_nine1)

Allele_freq = []

for line in three_nine:
    fields3 = line.strip('AF=')
    yes = fields3
    Allele_freq.append(yes)
    
df_freq = pd.DataFrame(list(zip(Allele_freq)), 
                  columns = ['Allele_freq'])
```
