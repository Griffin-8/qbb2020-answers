```python
import matplotlib.pyplot as plt
import pandas as pd
bed_columns = [ "chr", "start", "end"]
bed_columns
df_k4 = pd.read_table("../../qbb2020/data/K4me3.bed",
                     names=bed_columns)
df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]

df_k9 = pd.read_table("../../qbb2020/data/K9me3.bed",
                     names=bed_columns)

df_k9.loc[:,"width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]

df_k27 = pd.read_table("../../qbb2020/data/K27me3.bed",
                     names=bed_columns)

df_k27.loc[:,"width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
df_k9
```

```python

fig, ax = plt.subplots(nrows=3, figsize = (20,20))

#ax.bar( chr_k4.index, chr_k4 )

ax[0].hist(df_k4.loc[:,"width"], bins = 25)
ax[0].set_xlabel('Sequence Width')
ax[0].set_ylabel('Counts of Histone Methylation')
ax[0].set_title('K4 Methylation')

ax[1].hist(df_k9.loc[:,"width"], bins = 25)
ax[1].set_xlabel('Sequence Width')
ax[1].set_ylabel('Counts of Histone Methylation')
ax[1].set_title('K9 Methylation')
ax[1].set_xlim(0, 400000)
ax[1].set_ylim(0, 220)

ax[2].hist(df_k27.loc[:,"width"], bins = 25)
ax[2].set_xlabel('Sequence Width')
ax[2].set_ylabel('Counts of Histone Methylation')
ax[2].set_title('K27 Methylation')

plt.show()
```
