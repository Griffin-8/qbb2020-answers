```python
import matplotlib.pyplot as plt
import pandas as pd
bed_columns = [ "chr", "start", "end"]
bed_columns
df_k4 = pd.read_table("../../qbb2020/data/K4me3.bed",
                     names=bed_columns)
df_k4.loc[:,"width"] = df_k4.loc[:,"end"] - df_k4.loc[:,"start"]

df_k9 = pd.read_table("../../qbb2020/data/K9me3.bed",
                     names=bed_columns)

df_k9.loc[:,"width"] = df_k9.loc[:,"end"] - df_k9.loc[:,"start"]

df_k27 = pd.read_table("../../qbb2020/data/K27me3.bed",
                     names=bed_columns)

df_k27.loc[:,"width"] = df_k27.loc[:,"end"] - df_k27.loc[:,"start"]
```

```python

df_k4
```

```python
chr_k4 = df_k4.loc[:,"chr"]
chr_k4 = chr_k4.value_counts()


chr_k9 = df_k9.loc[:,"chr"]
chr_k9 = chr_k9.value_counts()


chr_k27 = df_k27.loc[:,"chr"]
chr_k27 = chr_k27.value_counts()

coi = ["2L", "2R", "3L", "3R","4", "X"]

```

```python

fig, ax = plt.subplots(nrows=3, figsize = (20,20))

#ax.bar( chr_k4.index, chr_k4 )

ax[0].bar(chr_k4.loc[coi].index, chr_k4.loc[coi])
ax[0].set_xlabel('Chromosome')
ax[0].set_ylabel('# of genes')
ax[0].set_title('K4 Chromosome Distribution')

ax[1].bar(chr_k9.loc[coi].index, chr_k9.loc[coi])
ax[1].set_xlabel('Chromosome')
ax[1].set_ylabel('# of genes')
ax[1].set_title('K9 Chromosome Distribution')

ax[2].bar(chr_k27.loc[coi].index, chr_k27.loc[coi])
ax[2].set_xlabel('Chromosome')
ax[2].set_ylabel('# of genes')
ax[2].set_title('K27 Chromosome Distribution')

plt.show()
```
