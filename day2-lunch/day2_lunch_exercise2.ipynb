```python
parsed = open('/Users/cmdb/qbb2020-answers/day2-lunch/parsed_mapping_file_2.txt')

D_parsed = {}

for line in parsed:
    fields = line.strip('\n') #could do .split to make fields into a list
    Fly_ID = fields[0:11]
    Uniprot = fields[12:]
    D_parsed[Fly_ID] = Uniprot

parsed.close()

```

```python
c_tab = open('/Users/cmdb/data/results/stringtie/SRR072893/t_data.ctab')

def identifier_mapping(c_tab, D_parsed, skip = ('pass') ):
    counter_1 = 0
    for line in c_tab:
        if 'FBgn' in line:
            split_line = line.split()
            FBGN = split_line[-4]
            if FBGN in D_parsed:
                split_line[-4] = D_parsed[FBGN]
                join_list="\t".join(split_line)
                if counter_1 < 100:
                    print(join_list)
                    counter_1 += 1
                else:
                    break
            else:
                skip
    print(counter_1)
    c_tab.close()

   
identifier_mapping(c_tab, D_parsed, skip = ('pass'))        
    

    
        
                       

    
        
        
```
